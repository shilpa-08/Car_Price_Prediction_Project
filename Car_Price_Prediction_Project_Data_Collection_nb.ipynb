# Collecting Data From olx site

```python
# importing necessary libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException , StaleElementReferenceException,ElementClickInterceptedException


# activating the chromedriver
driver=webdriver.Chrome(r'chromedriver.exe')
time.sleep(4)


# loading the webpage
url = "https://www.olx.in/"
driver.get(url)
time.sleep(4)
```

```python
# maximizing the window
driver.maximize_window()
time.sleep(4)
```

```python
# entering data in search_box
search_box= driver.find_element_by_xpath('//input[@data-aut-id="searchBox"]')
search_box.clear()
search_box.send_keys('Cars')
time.sleep(1)
```

```python

# clicking on the search button
search_button=driver.find_element_by_xpath('//div[@data-aut-id="btnSearch"]')
search_button.click()
```

```python
# code to reach till bottom of the page
for _ in range(1000):
    driver.execute_script("window.scrollBy(0,10000)")
    try:
        btn=driver.find_element_by_xpath('//button[@data-aut-id="btnLoadMore"]')
        btn.click()
        time.sleep(4)
    except NoSuchElementException:
        pass
```

```python
# getting links for car info
item_links=[]
for items in driver.find_elements_by_xpath('//li[@data-aut-id="itemBox"]/a'):  
    link=items.get_attribute('href')
    item_links.append(link)


```

```python
len(item_links)
```

```python
price=[]
brand_name=[]
model_name=[]
variant=[]
year=[]
fuel_type=[]
key_transmission=[]
km_driven=[]
no_of_owner=[]
location=[]
for i in item_links:
    driver.get(i)
    time.sleep(3)
    try:
        name=driver.find_element_by_xpath('//span[@data-aut-id="value_make"]')
        brand_name.append(name.text)
    except NoSuchElementException:
            brand_name.append(' - ')

    try:
            mname=driver.find_element_by_xpath('//span[@data-aut-id="value_model"]')
            model_name.append(mname.text)
    except NoSuchElementException:
            model_name.append(' - ')

    try:
            vrnt=driver.find_element_by_xpath('//span[@data-aut-id="value_variant"]')
            variant.append(vrnt.text)
    except NoSuchElementException:
            variant.append(' - ')

    try:
            yr=driver.find_element_by_xpath('//span[@data-aut-id="value_year"]')
            year.append(yr.text)
    except NoSuchElementException:
            year.append(' - ')


    try:
            fuel=driver.find_element_by_xpath('//span[@data-aut-id="value_petrol"]')
            fuel_type.append(fuel.text)
    except NoSuchElementException:
            fuel_type.append(' - ')

    try:
            transmission=driver.find_element_by_xpath('//span[@data-aut-id="value_transmission"]')
            key_transmission.append(transmission.text)
    except NoSuchElementException:
            key_transmission.append(' - ')

    try:
            km=driver.find_element_by_xpath('//span[@data-aut-id="value_mileage"]')
            km_driven.append(km.text)
    except NoSuchElementException:
            km_driven.append(' - ')

    try:
            owner=driver.find_element_by_xpath('//span[@data-aut-id="value_first_owner"]')
            no_of_owner.append(owner.text)
    except NoSuchElementException:
            no_of_owner.append(' - ')

    try:
            pr=driver.find_element_by_xpath('//span[@data-aut-id="itemPrice"]')
            price.append(pr.text)
    except NoSuchElementException:
            price.append(' - ')

    try:
            place=driver.find_element_by_xpath('//span[@class="_2FRXm"]')
            location.append(place.text)
    except NoSuchElementException:
            location.append(' - ')
    time.sleep(1)
```

```python
len(price)
```

```python
olx_cars=pd.DataFrame({})
```

```python
olx_cars['Brand']=brand_name
olx_cars['Model']=model_name
olx_cars['Variant']=variant
olx_cars['Manufacturing_year']=year
olx_cars['Driven_kilometers']=km_driven
olx_cars['Fuel']=fuel_type
olx_cars['number_of_owners']=no_of_owner
olx_cars['location']=location
olx_cars['Price']=price
```

```python
olx_cars.to_csv('Olx_car_data')
```
